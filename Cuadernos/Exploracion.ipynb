```python
## Pensemos en un experimento en el cual lanzamos una moneda y cada vez que sale cara multiplicamos por 2, cuando sale sello dividimos por 2.
## Vamos a hacer ese experimento 1000

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import shutil
np.random.seed(2025)

def simula(n_simulas=1000, n_pasos=100, valor_inicial=1, nombre_experimento="Proyecto1"):
    # cara= 1.1 y sello =1/1.1
    ruta_base= f"../datos/Experimentos/{nombre_experimento}/"
    if os.path.exists(ruta_base):
        shutil.rmtree(ruta_base)

    os.makedirs(ruta_base)
    valores_finales=[]
    for i in range(n_simulas):
        valor=valor_inicial
        trayectoria = []
        for j in range(n_pasos):
            trayectoria.append([j,valor])
            decision=np.random.uniform()
            #cara
            if decision <0.5:
                valor = valor*1.1
            else:
                valor = valor *(1/1.1)
    
        dataframe=pd.DataFrame(trayectoria)
        dataframe.columns=["Paso","Valor"]
        dataframe.to_csv(f"{ruta_base}/Experimento_{i}.csv",index=False)
        valores_finales.append(valor)

    return valores_finales     
        
vf=simula(n_simulas=1000,n_pasos=10,valor_inicial=1,nombre_experimento="proyecto_simula_2000")
```
