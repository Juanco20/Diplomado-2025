```python
from pathlib import Path    

def list_csv_files(directory: str) -> list:
    """List all Python files in the given directory."""
    p = Path(directory)
    return [str(file) for file in p.glob('*.py')]

listed_files = list_csv_files ('../ Datos')
```

```python
import logging

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

logging.info("Proceso iniciado")
archivo1=Path('../Datos/producos.csv')
if archivo1.exists():
    logging.info(f"El archivo {archivo1} existe.")
    logging.info(f"Tamaño del archivo: {archivo1.stat().st_size} bytes")
    logging.info(f"Nombre del archivo sin extensión: {archivo1.stem}")
    logging.info(f"Extensión del archivo: {archivo1.suffix}")
    logging.info(f"Directorio padre: {archivo1.parent}")
else:
    logging.warning(f"El archivo {archivo1} no existe.")
    logging.error("No se pueden obtener detalles del archivo inexistente.")
```

```python
import numpy as np
import logging

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

lista_elegibles= [1,2,3,0,0,"0","o","o","o","o"]

logging.info(f"Lista de elegibles: {lista_elegibles}")

## Escogemos dos


elegibles_1 = np.random.choice(lista_elegibles, size=2, replace=False)
logging.info(f"Elegibles 1: {elegibles_1}")

try:
    primer_elemento = int(elegibles_1[0])
    logging.info(f"Primer elemento seleccionado: {primer_elemento}")
    segundo_elemento = int(elegibles_1[1])
    logging.info(f"Segundo elemento seleccionado: {segundo_elemento}")
    
    div= primer_elemento / segundo_elemento
    logging.info(f"Resultado de la división: {div}")
except ZeroDivisionError:
    logging.error("Error: División por cero no permitida.")
except Exception as e:
    logging.error(f"Ocurrió un error inesperado: {e}")




 
```
